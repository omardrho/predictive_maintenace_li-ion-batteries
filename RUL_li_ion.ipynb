```python
X_train=df.iloc[:,0:1].values
y_train=df.iloc[:,5].values

test_data = pd.read_csv('testing_data.csv')
X_005=test_data.iloc[:,0:1].values
y_005=test_data.iloc[:,5].values
```

### Feature scaling

```python

ann = Sequential()
```

```python
ann.add(Dense(2 ,kernel_initializer= 'uniform',activation = 'tanh',input_dim = 1))
```

```python
ann.add(Dense(1,kernel_initializer= 'uniform',activation = 'linear'))
```

```python
ann.compile(optimizer = 'sgd', loss = 'mean_squared_error', metrics= ['mean_absolute_error','accuracy'])
```

```python
# Fitting the ANN in the Training set
ann.fit(X_train, y_train, batch_size = 1,epochs=2000)
```

```python
y_pred = ann.predict(X_005)
```
